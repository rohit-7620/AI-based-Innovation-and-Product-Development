# 9. Social Media Sentiment Trends for Financial Markets\n
\n
Build a sentiment trend signal from social media posts and compare with stock movement direction.\n

```python
import pandas as pd\n
from sklearn.feature_extraction.text import TfidfVectorizer\n
from sklearn.linear_model import LogisticRegression\n
\n
posts = pd.DataFrame({\n
    'text': [\n
        'Company earnings are excellent and growth is strong',\n
        'Bad guidance and weak quarter expected',\n
        'New product launch is promising',\n
        'Regulatory issues may hurt stock',\n
        'Analysts upgraded target price',\n
        'High debt levels are concerning'\n
    ],\n
    'sentiment': [1, 0, 1, 0, 1, 0]\n
})\n
\n
vec = TfidfVectorizer()\n
X = vec.fit_transform(posts['text'])\n
y = posts['sentiment']\n
clf = LogisticRegression(max_iter=1000).fit(X, y)\n
\n
sample = ['Earnings miss and weak demand outlook']\n
pred = clf.predict(vec.transform(sample))[0]\n
print('Predicted sentiment:', 'Positive' if pred == 1 else 'Negative')\n
```
