# 7. AI Product Architecture with Cloud + Edge Integration\n
\n
## Use Case\n
Smart retail shelf monitoring system.\n
\n
## Architecture Layers\n
1. Edge Layer: Camera + lightweight object detection for instant alerts\n
2. Gateway Layer: Device authentication and buffering\n
3. Cloud Layer: Central analytics, model registry, retraining pipeline\n
4. App Layer: Dashboard for operations and inventory teams\n

```python
architecture = {\n
    'edge': ['camera_stream', 'onnx_inference', 'local_alerts'],\n
    'gateway': ['iot_hub', 'device_auth', 'message_queue'],\n
    'cloud': ['data_lake', 'feature_store', 'model_training', 'model_registry'],\n
    'application': ['ops_dashboard', 'inventory_recommendations']\n
}\n
architecture\n
```

## Key Design Decisions\n
- Edge handles low-latency events\n
- Cloud handles heavy analytics and retraining\n
- MLOps pipeline supports periodic and drift-triggered model updates\n
