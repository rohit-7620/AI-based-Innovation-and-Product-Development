# 6. Building and Testing an AI Model\n
\n
## Application\n
Spam detection model using text messages.\n
\n
## Evaluation Metrics\n
- Accuracy\n
- Precision\n
- Recall\n

```python
import pandas as pd\n
from sklearn.model_selection import train_test_split\n
from sklearn.pipeline import Pipeline\n
from sklearn.feature_extraction.text import TfidfVectorizer\n
from sklearn.naive_bayes import MultinomialNB\n
from sklearn.metrics import accuracy_score, precision_score, recall_score\n
\n
df = pd.DataFrame({\n
    'text': [\n
        'Win cash now', 'Meeting at 5 pm', 'Limited offer claim prize',\n
        'Project update attached', 'Congratulations you won lottery', 'Let us have lunch tomorrow'\n
    ],\n
    'label': [1, 0, 1, 0, 1, 0]\n
})\n
\n
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.33, random_state=42)\n
model = Pipeline([\n
    ('tfidf', TfidfVectorizer()),\n
    ('clf', MultinomialNB())\n
])\n
model.fit(X_train, y_train)\n
pred = model.predict(X_test)\n
\n
print('Accuracy:', accuracy_score(y_test, pred))\n
print('Precision:', precision_score(y_test, pred, zero_division=0))\n
print('Recall:', recall_score(y_test, pred, zero_division=0))\n
```
