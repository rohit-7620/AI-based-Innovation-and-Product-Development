# 3. Sentiment Analysis for Customer Reviews\n
\n
This notebook builds a sentiment classifier for product/service reviews using TF-IDF + Logistic Regression.\n

```python
import pandas as pd\n
from sklearn.model_selection import train_test_split\n
from sklearn.feature_extraction.text import TfidfVectorizer\n
from sklearn.pipeline import Pipeline\n
from sklearn.linear_model import LogisticRegression\n
from sklearn.metrics import classification_report, accuracy_score\n
\n
data = pd.DataFrame({\n
    'review': [\n
        'Excellent product, highly recommended',\n
        'Terrible quality and bad support',\n
        'Good value for money',\n
        'Not worth the price',\n
        'Amazing service and fast delivery',\n
        'Very disappointed with the purchase'\n
    ],\n
    'label': [1, 0, 1, 0, 1, 0]\n
})\n
\n
X_train, X_test, y_train, y_test = train_test_split(\n
    data['review'], data['label'], test_size=0.33, random_state=42\n
)\n
\n
pipeline = Pipeline([\n
    ('tfidf', TfidfVectorizer(ngram_range=(1, 2))),\n
    ('clf', LogisticRegression(max_iter=1000))\n
])\n
pipeline.fit(X_train, y_train)\n
preds = pipeline.predict(X_test)\n
\n
print('Accuracy:', accuracy_score(y_test, preds))\n
print(classification_report(y_test, preds))\n
```

```python
new_reviews = [\n
    'The service was fantastic and smooth',\n
    'I regret buying this, very poor experience'\n
]\n
for r, p in zip(new_reviews, pipeline.predict(new_reviews)):\n
    print(r, '->', 'Positive' if p == 1 else 'Negative')\n
```
