```python
import tensorflow as tf
import gradio as gr
import numpy as np


model = tf.keras.applications.MobileNetV2(weights="imagenet")

def predict_fruit(img):
    # UX: Gradio handles the image upload, we just need to resize it
    img = tf.image.resize(img, (224, 224))
    img = tf.keras.applications.mobilenet_v2.preprocess_input(img)
    img = np.expand_dims(img, axis=0)

    # Inference
    preds = model.predict(img)
    decoded = tf.keras.applications.mobilenet_v2.decode_predictions(preds, top=3)[0]

    # Return as dictionary for the Gradio Label component
    return {label: float(conf) for (_, label, conf) in decoded}

# 2. Build the UI with Gradio
# This creates a drag-and-drop interface with a loading bar automatically.
demo = gr.Interface(
    fn=predict_fruit,
    inputs=gr.Image(),
    outputs=gr.Label(num_top_classes=3),
    title="üçì Smart Fruit Classifier",
    description="Upload a photo of a fruit (Apple, Banana, Orange, etc.)"
)

# 3. Launch
demo.launch(share=True)
```
