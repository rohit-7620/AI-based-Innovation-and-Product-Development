# 10. AI Deployment Techniques Comparison\n
\n
Compare cloud, on-premise, and edge deployment based on latency, scalability, and cost.\n

```python
import pandas as pd\n
\n
deployment_df = pd.DataFrame({\n
    'Environment': ['Cloud', 'On-Premise', 'Edge'],\n
    'Latency': ['Medium', 'Low-Medium', 'Lowest'],\n
    'Scalability': ['Very High', 'Medium', 'Low-Medium'],\n
    'Cost_Profile': ['OPEX, pay-as-you-go', 'High CAPEX + maintenance', 'Hardware distributed CAPEX']\n
})\n
deployment_df\n
```

```python
scores = pd.DataFrame({\n
    'Environment': ['Cloud', 'On-Premise', 'Edge'],\n
    'LatencyScore': [3, 4, 5],\n
    'ScalabilityScore': [5, 3, 2],\n
    'CostEfficiencyScore': [4, 2, 3]\n
})\n
scores['Total'] = scores[['LatencyScore', 'ScalabilityScore', 'CostEfficiencyScore']].sum(axis=1)\n
scores.sort_values('Total', ascending=False)\n
```

## Interpretation\n
- Cloud is best for fast scaling and managed operations\n
- On-premise fits strict control/compliance needs\n
- Edge is ideal for ultra-low latency and intermittent connectivity\n
